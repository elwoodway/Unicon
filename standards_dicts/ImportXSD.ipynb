```python
import re
from bs4 import BeautifulSoup
import pandas as pd
with open("./CoreMain.xsd") as fp:
    soup = BeautifulSoup(fp,'xml')

my_dict = {}
for element in soup.find_all('xs:element'):
    name = element['name']
    value = element.find('xs:documentation').get_text()
    my_dict[name] = value
```

```python
df = pd.DataFrame.from_dict(my_dict, orient='index')
```
